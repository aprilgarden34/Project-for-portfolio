```python
import os
import glob
```

```python
## 이름 (숫자).jpg
def rename_file(keyword):
    file_path = f'./downloads/{keyword}'
    files = os.listdir(file_path)

    for file in files:
        try:
            src = os.path.join(file_path, file)
            p = keyword
            m = ' (' + file.split('_')[1].split('.')[0] + ')'
            s = '.jpg'
            path = p+m+s
            dst = os.path.join(file_path, path)
            os.rename(src, dst)
        except:
            continue
```

```python
## 이름_숫자.jpg
def rename_file(keyword):
    file_path = f'./downloads/{keyword}'
    files = os.listdir(file_path)

    for file in files:
        try:
            src = os.path.join(file_path, file)
            p = file.split(' ')[0] + '_'
            m = file.split(' ')[1].split('.')[0]
            path = p+m+'.jpg'
            dst = os.path.join(file_path, path)
            os.rename(src, dst)
        except:
            continue
```
