```python
def get_plant_no(flower):
    url = 'http://openapi.nature.go.kr/openapi/service/rest/PlantService/plntIlstrSearch'
    params ={
        'serviceKey' : decoded_service_key, 
        'st' : '1', 
        'sw' : flower, 
        'dateGbn' : '', 
        'dateFrom' : '', 
        'dateTo' : '', 
        'numOfRows' : '20', 
        'pageNo' : '1' 
    }

    response = requests.get(url, params=params)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    search_li = soup.select('plantGnrlNm')
    for i in search_li:
        if i.text == flower:
            t = search_li.index(i)
            plant_no = soup.select('plantPilbkNo')[t].text
    try:
        return plant_no
    except:
        return False
```

```python
def get_plant_info(plant_no):
    url = 'http://openapi.nature.go.kr/openapi/service/rest/PlantService/plntIlstrInfo'
    params ={
        'serviceKey' : decoded_service_key,
        'q1' : plant_no
    }

    response = requests.get(url, params=params)

    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    name = soup.select_one('plantgnrlnm').text
    native = soup.select_one('dstrb').text
    flower = soup.select_one('flwrdesc').text
    if len(flower.strip()) == 0:
        flower = False
    return name, native, flower
```

```python
flower_info = {}
for flower in tqdm(flowers):
    dogam_no = get_plant_no(flower)
    if dogam_no:
        name, native, flower = get_plant_info(dogam_no)
        flower_info[dogam_no] = {
            'name':name,
            'native':native,
            'flower':flower
        }
```
