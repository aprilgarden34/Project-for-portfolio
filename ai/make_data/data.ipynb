```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./data/flowers.csv')
df.head()
```

```python
paldoLong=["전라남도","전라북도","경상남도","경상북도","충청남도","충청북도","경기도","강원도","제주도","전라남북도","경상남북도"]
paldoShort=["전남","전북","경남","경북","충남","충북","경기","강원","제주","전남 전북","경남 경북", "전국"]

df['refine_native']=' '

for i in range(len(df)):
    tmp=''
    for j in range(len(paldoLong)):
        if paldoLong[j] in df.loc[i]['native']:
            tmp+=paldoShort[j]+" "
    
    for j in range(len(paldoShort)):
        if paldoShort[j] in df.loc[i]['native']:
            tmp+=paldoShort[j]+" "
    
    #중복 제거
    tmp=' '.join(set(tmp.split(" "))).strip()

    if tmp!="":
        df.loc[i,'refine_native']=tmp
            
```

```python
# native 없는것 제거
empty=df[df['refine_native']==' '].index
print(empty)
df=df.drop(empty, axis='index', inplace=False)
```

```python
for d,c,r in zip(df['name'],df["native"], df['refine_native']):
    print(d)
    print(c)
    print(r)
    print(" ")
```
